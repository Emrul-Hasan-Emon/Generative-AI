# One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# Sample categorical data
data = np.array([['Red'], ['Green'], ['Blue'], ['Red'], ['Blue']])

# Initialize the OneHotEncoder
encoder = OneHotEncoder(sparse_output = False)  # sparse_output=False returns an array

# Fit and transform the data
encoded_data = encoder.fit_transform(data)
```

# Bag of Words (BoW)

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.DataFrame({"text":["people watch Emrul Hasan Emon",
                         "Emrul Hasan Emon watch many tutorials",
                         "people write comment",
                          "Emrul Hasan Emon write comment"],"output":[1,1,0,0]})

df
```

```python
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])
```

```python
bow
```

```python
bow.toarray()
```

```python
# Vocabulary
print(cv.vocabulary_)
```
